```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
km.fit(college.drop(['Unnamed: 0','Private'],axis=1))
```

```python
km.cluster_centers_
```

```python
def clust(x):
    if x == 'Yes':
        return 1
    else:
        return 0
```

```python
college['cluster'] = college['Private'].apply(clust)
```

```python
college
```

```python
km.labels_
```

```python
from sklearn import metrics
```

```python
print(metrics.confusion_matrix(college['cluster'],km.labels_))
print(metrics.classification_report(college['cluster'],km.labels_))
```
